 # Изучаем функцию: f(x) = 5x^3 - x^2 - 20x + 4
 
 ## Задачи: 
1. Определить корни
2. Найти интервалы, на которых функция возрастает
3. Найти интервалы, на которых функция убывает
4. Построить график
5. Вычислить вершину
6. Определить промежутки, на котором f > 0
7. Определить промежутки, на котором f < 0

**Приступаем к решению, импортируем библиотеки:**

```python
from sympy import *
from sympy.plotting import plot
init_printing()
```

**Прописываем функцию:**

```python
x = Symbol('x')
y = 5*x**3-x**2-20*x+4
y
```

**1. Определяем корни:**

```python
solve (y)
```

**2. Находим интервалы, на которых функция возрастает:**

```python
solve (diff(y)>0)
```

**3. Находим интервалы, на которых функция убывает:**

```python
solve (diff(y)<0)
```

**4. Строим график функции и ее производной в принципиальной области:**

```python
plot(y, diff (y), (x, -3, 3))
y, diff (y)
```

**6. Находим промежутки, на которых функция положительна:**

```python
solve (y>0)
```

**7. Находим промежутки, на которых функция отрицательна:**

```python
solve (y<0)
```

**5. Вычисляем минимумы и максимумы функции:**

из-за некорректной постановки задачи и очевидного поведения функции с учетом п.2<br/>
(отстутствия нижнего и верхнего предела) задача поиска минимума и максимума решается <br/>
на локальном участке [-2,2].

```python
array = solve (diff(y))
array
```

```python
x_max = float (array [0])
x_min = float (array [1])
```

```python
y_max = float(5*array[0]**3-array[0]**2-20*array[0]+4)
y_min = float(5*array[1]**3-array[1]**2-20*array[1]+4)
```

```python
max = [x_max, y_max]
min = [x_min, y_min]
```

```python
f'Координаты локального максимума {max}'
```

```python
f'Координаты локального минимума {min}'
```
